```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing the cars data
df = pd.read_csv("Cars93.csv")
print(df)
```

```python
# Filter the data for the specified manufacturers
selected_manufacturers = ["Audi", "Hyundai", "Suzuki", "Toyota"]
filtered_data = df[df["Manufacturer"].isin(selected_manufacturers)]

# Create a dictionary to store revs data for each manufacturer
revs_data = {manufacturer: filtered_data[filtered_data["Manufacturer"] == manufacturer]["Rev.per.mile"] for manufacturer in selected_manufacturers}

# Create a list of revs data for the specified manufacturers
data_to_plot = [revs_data[manufacturer] for manufacturer in selected_manufacturers]
print(data_to_plot)

# Create a box plot
plt.figure(figsize=(10, 6))
plt.boxplot(data_to_plot, labels=selected_manufacturers)
plt.title('Box Plot of Revs per Mile by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Revs per Mile (Rev.per.mile)')
plt.show()
```

Based on the boxplot Toyota is the manufacturer with a car that has the highest revs per mile followed by Suzuki, Hyundai and then Audi respectively

```python
# selecting columns
mpg_city = df['MPG.city']
mpg_highway = df['MPG.highway']

# creating histograms
plt.hist(mpg_highway,alpha=0.5,color="red",label="Mpg highway")
plt.hist(mpg_city,alpha=0.5,color="blue",label="Mpg city")

# adding labels and legend
plt.title('Distribution of MPG in the City and on the Highway')
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.legend()

```

Based on the graph above it is more fuel efficient to drive on the city than on the highway

```python
# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Wheelbase', y='Turn.circle', alpha=0.7)

# Fit a regression line (linear regression)
sns.regplot(data=df, x='Wheelbase', y='Turn.circle', scatter=False, color='red')

# Add labels and a title
plt.xlabel('Wheelbase (inches)')
plt.ylabel('Turning Circle (feet)')
plt.title('Relationship between Wheelbase and Turning Circle')
```

Based on the graph above there is a positive correlation between the wheelbase and the turning cycle. This suggests that as the wheelbase gets lager the turning cycle increases.


```python
# Group the data by car type and calculate the mean horsepower
mean_horsepower_by_type = df.groupby('Type')['Horsepower'].mean().reset_index()
print(mean_horsepower_by_type)
# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(mean_horsepower_by_type['Type'], mean_horsepower_by_type['Horsepower'], color='skyblue')

# Add labels and a title
plt.xlabel('Car Type')
plt.ylabel('Mean Horsepower')
plt.title('Mean Horsepower by Car Type')


```

Based on the above graph large does mean more horsepower on average

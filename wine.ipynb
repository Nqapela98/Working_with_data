```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("wine.csv")
del df["Unnamed: 0"]
selected_variety = ["Cabernet Sauvignon","Pinot Noir","Chardonnay"]
df_filtered = df[df["variety"].isin(selected_variety)]
```

```python
graph = sns.FacetGrid(data=df_filtered,col="variety",hue="variety")
graph.map(plt.hist,"points")
graph.set_ylabels("frequency")
```
